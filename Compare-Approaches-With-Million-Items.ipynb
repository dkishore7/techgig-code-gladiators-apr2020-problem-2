```python
"""
win10 functiion
window + searchsorted
"""
import numpy as np

def win10(list1, list2):
#     sl1 = sorted(list1)
    sl1 = np.sort(np.array(list1))
    sl2 = sorted(list2)

    list_1_min = min(sl1)
    list_1_max = max(sl1)

    list_2_min = min(sl2)    
    list_2_max = max(sl2)

    win_count =0

    idx = 0
    list_1_slider = 0
    list_1_max_index = len(sl1) -1

    while (idx < len(sl2)) and (sl2[idx] < list_1_max) and (list_1_slider <= list_1_max_index):
#         print(f'Testing for sl2 index {idx} for element {sl2[idx]} and list_1_slider {list_1_slider} part of sl1 {sl1[list_1_slider:]}')
        #result = list(map(lambda k: k > sl2[idx], sl1[list_1_slider:])).index(True)
        result = np.searchsorted(sl1[list_1_slider:],sl2[idx],side='right')
        if result <= list_1_max_index:
            win_count = win_count + 1
            idx = idx + 1
            list_1_slider = list_1_slider + result + 1
#             print(f'Found result {result} win_count {win_count} new_idx {idx} new_slider {list_1_slider}')

    print(win_count)
```

```python
import bisect

def win12(list1, list2):
	sl1 = sorted(list1)
	sl2 = sorted(list2)

	list_1_min = min(sl1)
	list_1_max = max(sl1)

	list_2_min = min(sl2)    
	list_2_max = max(sl2)
    
#     Look for obvious comparisons
	if list_2_max < list_1_min:
		win_count = len(sl1)
    
	elif list_1_max <= list_2_min:
		win_count = 0
    
	else:
		win_count =0

		idx = 0
		list_1_slider = 0
		list_1_max_index = len(sl1) -1

		while (idx < len(sl2)) and (sl2[idx] < list_1_max) and (list_1_slider <= list_1_max_index):
			result = bisect.bisect_right(sl1,sl2[idx],lo=list_1_slider,hi=list_1_max_index)
			if result > list_1_max_index:
				print('breaking')
				break
			else: 
				win_count = win_count + 1
				idx = idx + 1
				list_1_slider = result + 1
	
	print(win_count)
```
